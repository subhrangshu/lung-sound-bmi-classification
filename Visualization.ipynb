```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.3)
```

```python
os.listdir('../Data/ICBHI Processed Data')
```

```python
data = pd.read_csv('../Data/ICBHI Processed Data/Filtered_Processed_Labelled_Data.csv')
```

```python
data
```

```python
data = data.sample(frac=1).reset_index(drop=True)
#data = data.drop('Unnamed: 0', axis=1)
data = data.drop('Subject ID', axis=1)
data
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,6.2))
data[:100000].groupby('Category').boxplot(ax=axes, grid=True)
fig.suptitle('Distribution of respiratory auscultation data for the three categories of weights')
fig.savefig('Distribution.png', dpi=300)
```

```python
g=sns.pairplot(data.sample(frac=1).reset_index(drop=True)[:18081], hue='Category')
g.fig.set_size_inches(16,5)
g.fig.suptitle("Semi-processed respiratory auscultation data", y=1.08)
g.savefig('Semi-processed.png', dpi=200)
```

```python
data.groupby('Category').count()
```

```python
normalStd = data.groupby('Category').get_group('Normal').reset_index(drop=True)[:882000].groupby(data.groupby('Category').get_group('Normal').reset_index(drop=True)[:882000].index // 100).std()
normalStd['Category'] = 'Normal'
normalStd
```

```python
overweightStd = data.groupby('Category').get_group('Overweight').reset_index(drop=True)[:882000].groupby(data.groupby('Category').get_group('Overweight').reset_index(drop=True)[:882000].index // 100).std()
overweightStd['Category'] = 'Overweight'
overweightStd
```

```python
underweightStd = data.groupby('Category').get_group('Underweight').reset_index(drop=True)[:882000].groupby(data.groupby('Category').get_group('Underweight').reset_index(drop=True)[:882000].index // 100).std()
underweightStd['Category'] = 'Underweight'
underweightStd
```

```python
stdCombined = pd.concat([normalStd,overweightStd,underweightStd], axis=0)
stdCombined = stdCombined.reset_index(drop=True)
stdCombined = stdCombined.drop('Unnamed: 0',axis=1)
stdCombined
```

```python
stdCombined.to_csv('../Data/ICBHI Processed Data/Classification_ready.csv')
```

```python
f=sns.pairplot(stdCombined, hue='Category')
f.fig.set_size_inches(16,5)
f.fig.suptitle("V/s\nProcessed respiratory auscultation data", y=1.12)
f.savefig('Processed.png', dpi=200)
```
